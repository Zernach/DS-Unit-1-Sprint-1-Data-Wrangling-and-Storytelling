```python
import pandas as pd
pd.options.display.max_rows = 999

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows= 1, skipfooter = 2, engine = 'python')
print(df.shape)
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.drop(['id', 'desc', 'member_id', 'url'], axis = 1)
df.head()
```

```python
df.isnull().sum().sort_values()
```

```python
def int_rate_string_percentage_to_decimal(the_string):
  
  return float(float(the_string.strip().strip('%'))/100)

df['int_rate'] = df['int_rate'].apply(int_rate_string_percentage_to_decimal)

df.head()
```

```python
def term_string_to_int(the_string):
  
  return int(the_string.strip().strip(' months'))

df['term'] = df['term'].apply(term_string_to_int)

df.head()
```

```python
def numeric_loan_status_assignment(the_string):
  
  if the_string == 'Current' or the_string == 'Fully Paid':
    return 1
  
  else:
    return 0

  
df['loan_status_is_great'] = df['loan_status'].apply(numeric_loan_status_assignment)

df.head()
```

```python
df['loan_status_is_great'].value_counts()
```

```python
df['loan_status'].value_counts()
```

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format = True)
df['last_payment_d_year'] = df['last_pymnt_d'].dt.year
df['last_payment_d_month'] = df['last_pymnt_d'].dt.month

print(df['last_payment_d_year'].value_counts())
print(df['last_payment_d_month'].value_counts())
```
